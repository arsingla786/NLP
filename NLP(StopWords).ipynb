## **Stop Words**

Removing Words like -> I,the, those, what, why etc

```python
paragraph = "In the modern world, there are many people who are working in different fields, and they are trying to achieve their goals in life, but sometimes they do not realize that the path to success is not always easy and it requires a lot of patience and hard work. In order to succeed, one must be focused and should not give up when things are not going well, because there are many situations where a person may feel that they are not capable of doing something, yet with proper effort and determination, they can overcome those difficulties. Moreover, it is important to understand that learning is a continuous process and it does not stop at any stage of life, so people should always try to improve themselves and gain more knowledge, which can help them in the long run. At the same time, there are certain distractions that can affect productivity, and if a person is not careful, they might lose their direction and waste their valuable time. Therefore, it is essential to stay disciplined and maintain a proper balance between work and rest, as both are equally important for a healthy and successful life."
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words("english")
```

```python
#we can define our own stopords based on requirement
```

##Tokenization and removing stopwords

```python
#removing the stopwords form the paragraph
nltk.download('punkt_tab')
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
#tokenize tyhe para into sentences
sentences = nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
```

```python
##Apply stopwords and filter , and then apply stemming
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  #list comprehension
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words("english"))]
  sentences[i] = ' '.join(words) #joining all words to form new sentence
```

```python
sentences
```

```python
#try snowball stemmer for better stemming
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
##Apply stopwords and filter , and then apply stemming
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  #list comprehension
  words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words("english"))]
  sentences[i] = ' '.join(words) #joining all words to form new sentence

```

```python
sentences
```

##Lemmatization and removing Stopwords

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
##Apply stopwords and filter , and then apply lemmatization
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  #list comprehension
  words = [lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words("english"))]
  sentences[i] = ' '.join(words) #joining all words to form new sentence
```

```python
sentences
```

```python
#we see better results with lemmatization then stemming
```

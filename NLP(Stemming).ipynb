```python
#examples of stemming
#[goes,gone,going] -> go
#[eat, ate , eating ] -> eat
```

```python
words = [
    "playing", "played", "plays",
    "runner", "running",
    "happiness",
    "studies", "studying", "studied",
    "flying",
    "cries", "crying",
    "connected", "connecting", "connection", "connections",
    "better", "best",
    "walking", "walked", "walks",
    "quickly",
    "fairness", "kindness",
    "activation", "activated", "activating"
]
```

##Porter Stemmer

```python
import nltk
from nltk.stem import PorterStemmer
#this is a class so we have to initialize its object
```

```python
stemmer = PorterStemmer()
for word in words:
  print(word+"--->"+stemmer.stem(word))
#.stem(word) is an inbuilt function
```

```python
stemmer.stem('congratulations')
```

```python
stemmer.stem('sitting')
```

```python
#stemming works for good quantity of words but not for every word
#it might give bad result like in congratulations
```

image.png

```python
from nltk.stem import RegexpStemmer
regexstemmer = RegexpStemmer('ing$|s$|e$|able$', min=0)
print(regexstemmer.stem('happiness'))
# ing$ -> means that remove 'ing' from end
print(regexstemmer.stem("eating"))
```

##**SnowBall**  ***Stemmer*** (better than porter stemmer_)

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer("english")
for word in words:
  print(word + "-->"+snowballstemmer.stem(word))

```

```python
print(regexstemmer.stem("fairly"))  #will not give good result
print(snowballstemmer.stem("fairly"))  #will give good result
snowballstemmer.stem("history")
```

```python
#gives the exact root word , unlike root stem we get from stemming
#the output word is called LEMMA
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

lemmatize(word,pos)
\
**Pos = Nown - n(default),
Verb - v,
Adjective - a,
adverb - r**

```python
print(lemmatizer.lemmatize("going",pos = 'n'))
print(lemmatizer.lemmatize("going",pos = 'v'))
```

```python
words = [
    "playing", "played", "plays",
    "runner", "running",
    "happiness",
    "studies", "studying", "studied",
    "flying",
    "cries", "crying",
    "connected", "connecting", "connection", "connections",
    "better", "best",
    "walking", "walked", "walks",
    "quickly",
    "fairness", "kindness",
    "activation", "activated", "activating"
]
```

```python
for word in words:
  print(word + "----> "+ lemmatizer.lemmatize(word,pos = 'v'))
```

##Paragraph to Sentence (nltk.sent_tokenize)

```python
##paragraph->sentence
import nltk
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
nltk.sent_tokenize(corpus)
#gives a list of sentences from the paragaraph
```

```python
my_para = "Hi! How are you doing? Whats, going on."
nltk.sent_tokenize(my_para)
#separates with . ! ?
```

```python
my_para2 = "Artificial Intelligence (AI) is transforming the way humans live and work by enabling machines to perform tasks that typically require human intelligence. From healthcare and education to transportation and entertainment, AI is being used to improve efficiency, accuracy, and decision-making. For example, in healthcare, AI helps doctors diagnose diseases earlier and recommend better treatment plans, while in education, it enables personalized learning experiences for students."
```

```python
my_sentences = nltk.sent_tokenize(my_para2)
my_sentences
```

```python
type(my_sentences) #-> list
```

```python
for s in my_sentences:
  print(s)
```

##Paragraph to Words (ntlk.word_tockenize)

```python
#paragraph/sentences to words or , sentence to words
from nltk.tokenize import word_tokenize
my_para3 = "AI can help doctors find diseases and help students learn better. It makes work faster and easier, but it also has some risks like loss of jobs and privacy issues."
words = nltk.word_tokenize(my_para3)
words
```

```python
type(words) #->list
```

```python
#it also prints . , ! and symbols
```

```python
document = sent_tokenize(my_para3)
for s in document:
  print(word_tokenize(s))
#prints list of words in each sentence
```

##wordpunct_tokenize

```python
#wordpunct_tokenize -> punctautions like 's  are also treated separately
from nltk.tokenize import wordpunct_tokenize
test_para = "This is akriti's phone."
wordpunct_tokenize(test_para)
```

##word bank tokenize -> (.) is not treated as a separate word
